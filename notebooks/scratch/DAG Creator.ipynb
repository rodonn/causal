```python
import graphviz as gr
```

```python
g = gr.Digraph()
g.edge('A','B')
g.edge('B','C')
g.edge('C','A')
g.edge('C','D')
g.edge('D','B')
g.edge('D','A')
g.render('dags/cyclic', format='png')
g
```
